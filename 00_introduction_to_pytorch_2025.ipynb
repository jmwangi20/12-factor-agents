## Random Tensors

### Why random tensors?
Random tensors are very important in the neural networks because of the way they learn.They ussually start with random numbers and adjust gradually as they learn.

```start with random numbers -> look at the data -> update the random numbers -> look at the data```

````python
#create a random tensor of size (3,4)
random_tensor = torch.rand(3,4)
random_tensor
````

````python
#check the dimension of the random tensor
random_tensor.ndim
````

````python
#create an image size tensor

random_image_size_tensor = torch.rand(size = (224,224,3))
random_image_size_tensor.shape,random_image_size_tensor.ndim
````

````python
 #create tensors of zeros and ones
 zeros = torch.zeros(size=(3,4))
 zeros.dtype,zeros.ndim,zeros.shape
````

````python
#create tensors of ones
ones = torch.tensor(size = (3,4))
ones.dtype,ones.ndim,ones.shape
````

### Creating a range of tensors and tensorls-like

````python
#use torch.arange to create range tensors
zero_to_ten = torch.arange(start=0,end=11,step=1)
zero_to_ten
````

````python
# tensors_like create tensors of the same shape as of the like to tensor
ten_zeros = torch.zeros_like(zero_to_ten)
ten_zeros
````

## Tensor Datatypes
**Note:** Tensors datatypes is one of the big errors you`ll run into with pytorch and Deeplearning.

  1. Tensors not in the right datatype

  2. Tensors not in the right shape

  3. Tensors not in the right device



````python
# float_32_tensors
float_32_tensor = torch.tensor([3.0,6.0,9.0],
                               dtype=None,
                               device=None,
                               requires_grad=False)
float_32_tensor
````

````python
# Convert float32 into float16 datatype
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor

````

````python
#Getting tensor attributes

some_tensor = torch.rand(3,4)
some_tensor
````

````python
#print the tensor attributes
print(some_tensor)
print(f"Data type of the tensor:{some_tensor.dtype}")
print(f"shape of the tensor:{some_tensor.shape}")
print(f"Device the tensor is on:{some_tensor.device}")
````
